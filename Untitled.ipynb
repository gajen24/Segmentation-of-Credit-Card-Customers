# Deriving New KPI



1. Monthly average purchase and cash advance amount

```python
credit['Monthly_avg_purchase']=credit['PURCHASES']/credit['TENURE']
```

```python
credit['Monthly_cash_advance']=credit['CASH_ADVANCE']/credit['TENURE']
```

```python
print(credit['Monthly_avg_purchase'].head())
print(credit['Monthly_cash_advance'].head())
```



2- Purchases by type (one-off, installments)

```python
credit[credit['ONEOFF_PURCHASES']==0]['ONEOFF_PURCHASES'].count()
```

```python
credit.loc[:,['ONEOFF_PURCHASES','INSTALLMENTS_PURCHASES']]
```

```python
print(credit[(credit['ONEOFF_PURCHASES']==0) & (credit['INSTALLMENTS_PURCHASES']==0)].shape)

print(credit[(credit['ONEOFF_PURCHASES']>0) & (credit['INSTALLMENTS_PURCHASES']>0)].shape)

print(credit[(credit['ONEOFF_PURCHASES']>0) & (credit['INSTALLMENTS_PURCHASES']==0)].shape)

print(credit[(credit['ONEOFF_PURCHASES']==0) & (credit['INSTALLMENTS_PURCHASES']>0)].shape)
```

```python
def purchase(credit):
    if (credit['ONEOFF_PURCHASES']==0) & (credit['INSTALLMENTS_PURCHASES']==0):
        return 'none'
    if (credit['ONEOFF_PURCHASES']>0) & (credit['INSTALLMENTS_PURCHASES']>0):
         return 'both_oneoff_installment'
    if (credit['ONEOFF_PURCHASES']>0) & (credit['INSTALLMENTS_PURCHASES']==0):
        return 'one_off'
    if (credit['ONEOFF_PURCHASES']==0) & (credit['INSTALLMENTS_PURCHASES']>0):
        return 'istallment'
```

```python
credit['purchase_type']=credit.apply(purchase,axis=1)
```

```python
credit['purchase_type'].value_counts()
```

# Clustering

```python
from sklearn.cluster import KMeans
```

```python
# selecting the number of clusters
wcss=[]

for i in range(1,20):
    clusters=KMeans(i)
    clusters.fit(reduced_cr)
    wcss_iter=clusters.inertia_
    wcss.append(wcss_iter)
```

```python
wcss
```

```python
number_cluster=range(1,20)
plt.plot(number_cluster,wcss,marker = "o")
plt.title('The Elbow Methode')
plt.xlabel('Number of cluster')
plt.ylabel('Within-cluster sum of squares')
```

From above graph we will find elbow range. here it is 4,

# Silhouette Coefficient

```python
from sklearn import metrics
```

```python
k_range = range(2, 7)
scores = []
for k in k_range:
    km = KMeans(n_clusters=k, random_state=1)
    km.fit(reduced_cr)
    scores.append(metrics.silhouette_score(reduced_cr, km.labels_))
    
scores    
```

```python
plt.plot(k_range, scores,marker = "o")
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Coefficient')
plt.grid(True)
```

so it's conform that no of cluster=4, that we are taking.
